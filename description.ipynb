# Ollama simple Chatbot APP

# Ollama model

- Ollama allows you to run LLMs locally (like LLaMA, Mistral, Gemma) on your own machine without sending data to the cloud. When integrated with LangChain, Ollama acts as a drop-in LLM provider similar to Gemini or OpenAI.

# Langsmith

- LangSmith, on the other hand, is an observability and evaluation platform for LLM applications. It tracks prompts, responses, latency, errors, and chain behavior.
